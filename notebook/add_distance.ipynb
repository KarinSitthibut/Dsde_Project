```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(
    "../data/processed/footpath_phase1.csv",
    on_bad_lines="skip",
    engine="python"
)

offices = pd.read_csv(
    "../data/external/district_offices.csv"
)
```

```python
def haversine(lat1, lon1, lat2, lon2):
    R = 6371000  # meters
    lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2])
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = np.sin(dlat/2)**2 + np.cos(lat1)*np.cos(lat2)*np.sin(dlon/2)**2
    c = 2 * np.arcsin(np.sqrt(a))
    return R * c
```

```python
dist_matrix = []

for _, office in offices.iterrows():
    d = haversine(
        df["lat"],
        df["lon"],
        office["lat"],
        office["lon"]
    )
    dist_matrix.append(d)

dist_matrix = np.vstack(dist_matrix).T
dist_matrix.shape
```

```python
df["distance_to_office_m"] = dist_matrix.min(axis=1)
df.head()
```

```python
df.to_csv(
    "../data/processed/footpath_with_distance.csv",
    index=False
)

print("Saved to footpath_with_distance.csv")
```
